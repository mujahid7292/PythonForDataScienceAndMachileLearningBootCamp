```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x='sex',data=tips)
```

# sns.set_style()

```python
# Dark BackGround
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# White BackGround
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
# Ticks on the edge of the graph
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

# Remove Graph's Spine

```python
# Remove Graph's Upper & Right Spine
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
# Remove Graph's All The Spine
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

# Size & Aspect

We can resize & change `seaborn`'s graph using `matplotlib`'s  <b>plt.figure(figsize=(W,H))</b> command. As under the hood `seaborn` actually talks with `matplotlib`

```python
# Set graph width=12 inches, height = 3inches
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

# Scale & Context

```python
# Lets say we want to output figure for our poster. Then we will set_context = poster
plt.figure(figsize=(10,5))
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context(context='poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

# Palette & Color

```python
sns.set_context(context='notebook',font_scale=1)
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='viridis')
```
