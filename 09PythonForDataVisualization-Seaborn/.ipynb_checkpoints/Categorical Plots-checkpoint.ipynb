```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

Barplot aggregate catagorical data based on some function. Default function is mean. Here barplot is showing mean number of bill based on sex. That means mean amount of bill male customer get is roughly 21 dollar & mean amount of bill female customer get is roughly 18 dollar.

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

This bar plot is showing standard deviation (that means variation) on the mean bill amount based on male and female. We are seeing that there is much higher variation in male bill than female bill amount.

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.std)
```

# Count Plot

```python
sns.countplot(x='sex',data=tips)
```

Count plot will show the number of occurence in per catagory. In our tips dataset we have 244 rows of data. Among the roughly 160 bill is paid by male customer & roughly 84 bill paid by female customer.

# Box Plot

box plot's nice explanation in youtube: <br/><br/>
https://www.youtube.com/watch?v=BhHOdKyFzyQ  <br/><br/>

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

`Smoker tend to spend more in weekend (saturday & sunday) than non smoker.`

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips,hue='day')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

Male spend lot in friday night compare to weekend. On the other hand female spend less in friday night but catch up male in weekend.

# Swarm Plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

# Combine Violin & Swarm Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,color='white')
sns.swarmplot(x='day',y='total_bill',data=tips)
```
