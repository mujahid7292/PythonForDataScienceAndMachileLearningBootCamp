### Functional Method:

###### Single Plot In Single Canvas

```python
# (xAxix, yAxix, color)
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sample Plot')
```

```python
# (#of Rows, #of Columns, plotNumber)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure Object (Figure is empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # Left, Bottom, Width, Height (range 0 To 1)

# Now we will plot on this axis
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title('Title')
```

# Setting Two Axes/Chart In Single Canvas / Figure

```python
# Create a figure object (figure is our empty canvas)
fig = plt.figure()

# Create two chart. This two chart will be in the same figure / canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # This is the big chart
axes1.set_title('Big Chart')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.plot(x,y)

axes2 = fig.add_axes([ 0.20 , 0.50 , 0.40, 0.30 ]) # This is the small chart
axes2.set_title('Small Chart')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.plot(y,x)
```

#### Explanation Of The Axes Value:
Big Charts Axes (axes1) = [ 0.1 , 0.1 , 0.8, 0.8 ]

0.10 = axes1 start 10% from the left of the canvas.<br/>
0.10 = axes1 start 10% from the bottom of the canvas.<br/>
0.80 = axes1 take 80% of the canvas width.<br/>
0.80 = axes1 take 80% of the canvas height.<br/><br/>

Small Charts Axes (axes2) = [ 0.20 , 0.50 , 0.40, 0.30 ]<br/>

0.20 = axes2 start 20% from the left of the canvas.<br/>
0.50 = axes2 start 50% from the bottom of the canvas.<br/>
0.40 = axes2 take 40% of the canvas width.<br/>
0.30 = axes2 take 80% of the canvas height.<br/><br/>

```python
# Axes is an array of axes to plot on
axes
```

```python
# So We can use for each on 'axes' array
for chart in axes:
    chart.set_title('Title')
    chart.set_xlabel('X Label')
    chart.set_ylabel('Y Label')
fig
```

```python
# As axes are bassically index, so we can grab axes by it's index number.
fig,axes1 = plt.subplots(nrows=1,ncols=3)
axes1[0].plot(x,y)
axes1[0].set_title('Chart-1')

axes1[1].plot(y,x)
axes1[1].set_title('Chart-2')

axes1[2].plot(y**3,x)
axes1[2].set_title('Chart-3')

plt.tight_layout()
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(3,4),dpi=200)
# Here fisize=(3 inches wide, 4 inches height)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

We can also set figSize and dpi to our subplot.

```python
fig,axes = plt.subplots(figsize=(2,1), dpi=200, nrows=1, ncols=2)
# Here fisize=(2 inches wide, 1 inches height)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, label="X And Y")
ax.plot(y,x, label="Y And X")
ax.legend()
```

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details. Some of the most common **loc** values are:

```python
# Lots of options....

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```
