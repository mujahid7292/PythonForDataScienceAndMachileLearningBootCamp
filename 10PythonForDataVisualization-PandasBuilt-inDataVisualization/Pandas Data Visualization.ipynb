# Scatter Plot

```python
df1.plot.scatter(x='A',y='B')
```

Above we are seeing for corrosponding A value what is the B value. If we want we can add another column in scatter plot. But we have to bring this another column as a size argument.

```python
df1.plot.scatter(x='A',y='B',s=df1['C'])
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100) # scale the plot by 100 
```

So where C value is big points are big, where C value is small points are small.

# HexBin Plot

For this hexbin plot example create a dummy bivariate (2 variables) dataFrame below.

```python
df = pd.DataFrame(data=np.random.randn(1000,2),columns=['A','B'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='A',y='B')
```

```python
df.plot.hexbin(x='A',y='B',gridsize=30)
```

```python
df.plot.hexbin(x='A',y='B',gridsize=30,cmap='coolwarm')
```
